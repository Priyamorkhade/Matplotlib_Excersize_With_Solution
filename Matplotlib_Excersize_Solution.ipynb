#Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

'''
## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**
'''

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel(" X Axes")
ax.set_ylabel(" Y Axes")
ax.set_title("Graph Of X Vs Y")
```

Notice the xlimits and y limits on the inserted plot:**


```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z,'b')
ax1.set_xlabel(" X")
ax1.set_ylabel(" Z")


ax2.plot(x,y,'b')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title("zoom")
plt.show()
```

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. 
Play around with the linewidth and style**

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color="blue",ls="--",lw="2.5")
axes[1].plot(x,z,'r',linewidth="2.5")
plt.show()
```


** See if you can resize the plot by adding the figsize() argument

```python
fig,axes=plt.subplots(1,2,figsize=(15,3))
axes[0].plot(x,y,'b',lw="6")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x,z,'r--',lw=4)
```
